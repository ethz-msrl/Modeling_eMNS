```python
import numpy as np
import pandas as pd
import sys
sys.path.append("../")
from functions.functions import load_data_forGridSearch
```

```python
# load training data
X_train, y_train = load_data_forGridSearch("../../Data", "train")
```

```python
train = pd.DataFrame(np.hstack((X_train, y_train)), columns = ["x", "y", "z", 
                                 "I1", "I2","I3", "I4","I5", "I6","I7", "I8",
                                 "Bx", "By", "Bz"])
train.head()
```

```python
def filter_by_max_currents(df, max_currents=5):
    
    # filtered = df[df[3:-3]<=max_currents and df[3:-3]>=-max_currents]
    # filtered = df.loc[(-max_currents<=df[3:-3]<=max_currents).all()]
    # filtered = df.loc[(-max_currents<=df[3:-3]) & (df[3:-3]<=max_currents)]
    # filtered = df.drop()
    filtered = df[(df.I1 <= max_currents) & (-max_currents <= df.I1) &
                  (df.I2 <= max_currents) & (-max_currents <= df.I2) &
                  (df.I3 <= max_currents) & (-max_currents <= df.I3) &
                  (df.I4 <= max_currents) & (-max_currents <= df.I4) &
                  (df.I5 <= max_currents) & (-max_currents <= df.I5) &
                  (df.I6 <= max_currents) & (-max_currents <= df.I6) &
                  (df.I7 <= max_currents) & (-max_currents <= df.I7) &
                  (df.I8 <= max_currents) & (-max_currents <= df.I8)]
    # filtered = filtered.dropna()
    
    return filtered
```

```python
train_filtered = filter_by_max_currents(train)
```
